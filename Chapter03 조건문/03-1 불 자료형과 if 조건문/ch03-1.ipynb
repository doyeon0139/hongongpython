> bool 연산하기 : 논리 연산자

```python
# bool 끼리 연산 : 논리 연산자(not, and, or) && or ||

# and 는 All 모든 값이 True 이어야 True 반환
True and True
True and True and True and True and True and True and True and True and True and False

# or 는 any 어느 하나라도 True이면 True 반환
True or False
False or False or False or False or False or True

# not 은 논리값 뒤집기(반대)
not True
not False

# 논리 연산자 우선순위 : not > and >or
not True and False or not False
#  False and False or True
#       False     or  True
#            True


```

> 논리 연산자와 비교 연산자를 함께 사용

```python
# 비교 연산자로 비교한 결과를 논리 연산자로 최종 리턴
10 == 10 and 10 != 5
10 > 5 or 10 < 3
not 10 > 5
```

```python
# print(1 + True)  # = 2 ??

#0, 0.0 이외의 모든 숫자는 True
bool(1234)
bool(-1)
bool(0)
bool(3.14)
bool(0.0)

print(1 + False)

# "" 제외한 모든 문자열 True
bool('문자열')  #True
bool('')   #False

# {}, [], () 등 빈 컨테이너(리스트, 튜플, 딕셔너리 등의 값이 비어 있는 경우) False 제외한 모든 자료형 True
bool([1, 2, 3])
bool([])
bool({1:a, })

```

> 단락 평가(short-circuit evalu) : 표현식에 대한 평가를 생략하는 방법

```python
# 첫 번째 값이 거짓이면 두 번째 값은 확인하지 않고 바로 거짓으로 결정
False and True
10 < 1 and 3 == 3 and 4 % 2 == 0

# 첫 번째 값이 참이므로 두 번째 값은 확인하지 않고 참으로 결정
True or True
True or False
10 < 1 or 3 == 3 or 4 % 2 == 0

# 논리 연산식을 연결할 때 논리 판단 표현식의 순서에 따라서 프로그램의 성능에 영향을 준다.
```

> bool 의 사용 : if, while 등 조건식을 만들 때 사용

```python
if True :
    print('참')

n = 13
if n % 2 == 0:
    print(f'{n}은 짝수 입니다.')
else :
    print(f'{n}은 홀수 입니다.')

li = []
print(li)
print(bool(li))

if not li:
    li.append(1)

print(li)

if li:

    li.append(2)

print(li)
```

```python
# 프로그램에서 참과 거짓의 쓰임새 : while

if [1, 2, 3]:
    print('참')
else :
    print('거짓')

a = ['a','b','c','d','e','f']
print(a)

while a:
    a.pop()
    print(a)
```

> if 조건문 : 조건에 따라 코드의 실행을 분기 하고자 할 때(실행의 흐름을 제어)

```python
score = 55
CUT_LINE = 60
if score >= CUT_LINE:
    print('합격 하셨습니다.')
    print('축하드립니다.')
else:
    print('땡 불합격입니다.')
```

```python
# 사용자로부터 숫자를 입력받아 양수, 음수를 판별하는 프로그램 작성

num = (input('숫자를 입력하세요'))

if num.isdecimal :
    num = int(num)
    if num > 0 :
        print('양수입니다.')

    elif num < 0:
        print('음수입니다.')

    else : 
        print('0 입니다.')

else:
    print('잘못 입력했습니다.')
```

```python
input_data = input('정수를 하나 입력하세요')
number = 0

# 입력데이터 유효성검증

# 숫자라면...
if input_data.isdigit:
    number = int(input_data)
    print('입력값 : ',number)

# 숫자가 아니라면..
else:
    print('숫자가 아닙니다')
    print('프로그램을 종료합니다.')
    # quit()
    
# 양수/음수 판별
if number > 0:
    print('양수입니다.')

elif number < 0:
    print('음수입니다.')

else:
    print('0 입니다.')
```

```python
try:
    dat = int(input())
    if dat > 0:
        print('+')
    elif dat < 0:
        print('-')
    else:
        print('0')
except:
    print('정수만 입력해주세요')
```

```python
dat = input()
if dat.lstrip('+-').isdigit() and dat.count('+')<2 and dat.count('-')<2:
    dat = int(dat)
    if dat > 0:
        print('+')
    elif dat < 0:
        print('-')
    else:
        print('0')
else:
    print('입력값이 정수가 아닙니다.')
```

```python
num = input("숫자를 입력하세요: ")

Pass = False

if num[0] == "-" :
    Pass = True
elif num.isnumeric() :
    Pass = True

while Pass == False :
    num = input("실수(정수)가 아닙니다, 숫자를 입력하세요: ")
    if num[0] == "-" :
        Pass = True
    elif num.isnumeric() :
        Pass = True

if float(num) > 0:
    print('양수입니다')
elif num[0] == "-" :
    print('음수입니다')
else:
    print('0입니다')
```
