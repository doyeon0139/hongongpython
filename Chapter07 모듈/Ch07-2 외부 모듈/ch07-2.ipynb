> BeautifulSoup 모듈 이용한 웹 스크랩핑

```python
# 웹 페이지 분석 모듈

# 현재위치와 현재의 온도 출력
# 네이버 오늘의 날씨(https://weather.naver.com/today/02281112)

from urllib import request
from bs4 import BeautifulSoup

target = request.urlopen('https://weather.naver.com/today/02281112')
soup = BeautifulSoup(target, 'html.parser')

# 현재위치
data = soup.find('div',{'class':"location_area"})
addres = data.find('strong', {'class':'location_name'}).text

wheather = soup.find('strong',{'class':'current'}).get_text().replace("현재 온도",'')


print('오늘 {}의 날씨는 {}입니다.'.format(addres,wheather))
```
