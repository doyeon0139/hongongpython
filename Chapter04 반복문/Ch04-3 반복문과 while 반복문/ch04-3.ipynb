```python
# 1. for looping : íŠ¹ì • íšŸìˆ˜ ë§Œí¼ ë°˜ë³µí•˜ëŠ” ê²½ìš°(ë°˜ë³µ íšŸìˆ˜ê°€ ì •í•´ì ¸ ìˆì„ ë•Œ)
# 2. while looping : íŠ¹ì • ì¡°ê±´ì— ëŒ€í•˜ì—¬ ë°˜ë³µí•˜ëŠ” ê²½ìš°(ì¡°ê±´ì‹ì´ ë§Œì¡±í•  ë•Œê¹Œì§€ ë°˜ë³µ)
# while boolí‘œí˜„ì‹ : 

# íšŸìˆ˜ ë°˜ë³µ
import time
for i in range(15):
    print(i, end=" ")
    time.sleep(0.5)
print('for ë°˜ë³µë¬¸ ì¢…ë£Œ')

# ì¡°ê±´ ë°˜ë³µ
n = 0
while n < 15:
    print(n, end=" ")
    n += 1
    time.sleep(0.5)
print('while ë°˜ë³µë¬¸ ì¢…ë£Œ')
```

```python
import string
li = """!"P#$%&'y()*+,t-./:h;<=o>?@[\]n^_`{|}~"""
ans = []

l = len(li)
for i in range(l):
    if li[i] not in string.punctuation:
        ans.append(li[i])

print(''.join(ans))
```

```python
import string
sen = list("""!"P#$%&'y()*+,t-./:h;<=o>?@[\]n^_`{|}~""")
ans = []
while sen:
    tmp = sen.pop(0)
    if tmp not in string.punctuation:
        ans.append(tmp)
print(''.join(ans))
```

```python
import string

signs = """!"P#$%&'y()*+,t-./:h;<=o>?@[\]n^_`{|}~"""
word = ''
for ch in signs:
    if ch not in string.punctuation:
        word += ch

print('word : ', word)
```
