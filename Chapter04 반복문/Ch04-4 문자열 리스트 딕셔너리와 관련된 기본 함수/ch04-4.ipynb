> 리스트에 적용할 수 있는 기본함수 : main(), max(), sum()

```python
# 1 ~ 10 까지의 합
li = list(range(1, 10 + 1))
sum(li)

scores = [96, 87, 98, 63, 56]
print('최고점 : ', max(scores))
print('최저점 : ', min(scores))
print('총 점 : ', sum(scores))
print('평 균 : ', sum(scores)/(len(scores)))

print(type(scores.sort()))

scores.sort # 요소가 오름차순 정렬
print(scores)
print("최고점 : ", scores[-1])
print("최저점 : ", scores[0])

```

> enumerate() 함수와 반복문 조합하기

```python
# enum : 열거형 상수
# 부서코드 : 관리부(012)...

alpas = ['a', 'b', 'c', 'd']

enum_li = enumerate(alpas)
print('enumerated{} = {}'.format(alpas, enum_li))
print('enumerated{} = {}'.format(alpas, list(enum_li)))

# i = int(i)

# for i in (len(alpas)):
#     print(i, alpas[i])

for i, e in enumerate(alpas):
    print(i, '=', e)
```

> 딕셔너리의 items() 함수와 반복문 조합하기

```python
di = {"A+":95, "A":90, "B+":85, "B":80, "D":60}
for e in di:
    print(e) 

print(di.items())

for k, v in di.items():
    print('키:{}, 값{}'.format(k, v))
```

> 리스트 내포(list comprehension)

```python
# [] 리스트 안에 식, for 반복문, if 조건문 등을 지정하여 리스트를 생성하는 것
# {식 for 변수 in 반복자료형 if 조건식}
# list(식 for 변수 in 반복자료형 if 조건식)

li_1 = list(range(10)) #0~9까지
print('li_1 = ', li_1)

li_2 = [i for i in range(10)]
print('li_2 = ', li_2)

li_2 = [i*100 for i in range(10)]
print('li_2 = ', li_2)

print()

li_3 = [i**i for i in range(10)]
print("li_3 =", li_3)

li_3_1 = []
for i in range(10):
    li_3_1.append(i**i)


# 짝수 리스트 생성
li_5 = []
for i in range(10):
    if i % 2 == 0:
        li_5.append(i)
print('li_5 = ', li_5)


li_5_1 = [i for i in range(10) if i % 2==0 ]
print('li_5_1 = ', li_5_1)
```

```python
presidents = ['이승만', '박정희', '전두환', '최규하', '노태우', '김영삼', '김대중', '노무현', '이명박', '문재인']

# 성이 '김', '이', '박' 인 대통령을 추출하여 리스트를 생성
output = [i for i in presidents if i[0] in '이김박']
print('output =', output)
```
