> 리스트에 요소 제거하기 : 인덱스로 제거, 값으로 제거

```python
# 인덱스 사용
li = ['a','b','c','d','e','f']

# 1. del
# del li
print(li) #nameError

del li[1]
print(li)

del li[:2]
print(li)

# 2. 리스트.pop(인덱스)
print('삭제요소:', li.pop(2))
print(li)
print('삭제요소:', li.pop())
print(li)
```

```python
# 값으로 제거

# 리스트 list.remove(값)
li = [1,2,1,2]
print(li)

li.remove(2) # 리스트 내부에 해당하는 값이 여러개 있어도 가장 먼저 발견되는 하나만 제거
print(li)

li.remove(2) 
print(li)

```

```python
# 모두 제거 : clear()  del과 차이점 알아두기
li = [0,1,2,3,4,5]
print(li)

li.remove(0)
print(li)

li.clear()  
print(li)
```

> 아래의 성적 레코드의 최고, 최저, 총점, 평균을 구하는 프로그램 작성 : record.py

```python
records = [90, 85, 70, 55, 95, 60, 80, 90, 85, 70, 55, 95, 60, 80, 90, 85, 70, 55, 95, 60, 80]
# len(records)

sum = 0
avg = 0
max = records[0]
min = records[0]
n = 0

for n in range(len(records)):
    if max < records[n] :
        max = records[n]
    if min > records[n] :
        min = records[n]
    n += 1

for i in records :
    sum += i
    
avg = sum / len(records)

print('최고점 : ', max)
print('최저점 : ', min)
print('총  점 : ', sum)
print('평  균 : ', avg)
```

```python
# import random
# records = [random.randrange(0, 99) for _ in range(10)]
records = [90, 85, 70, 55, 95, 60, 80, 90, 85, 70, 55, 95, 60, 80, 90, 85, 70, 55, 95, 60, 80]
max_score = 0
min_score = 100
total_score = 0
for score in records:
    if max_score < score:
        max_score = score
    if min_score > score:
        min_score = score
    total_score += score
print("점수 목록: ", records)
print(f"최고:{max_score}, 최저:{min_score}, 총점:{total_score}, 평균:{total_score/len(records) : 0.2f}")
```

```python
records = [90, 85, 70, 55, 95, 60, 80, 90, 85, 70, 55, 95, 60, 80, 90, 85, 70, 55, 95, 60, 80]
len(records)


a = max(records[:])
b = min(records[:])

print(f"최고점: {a}" )
print(f"최저점: {b}")

sum = 0
for i in records:
    sum += i
print(f"총점: {sum}")


print(f"평균: {sum/len(records)}")
```

> 아래와 같이 실행 결과가 나오는 프로그램 작성 : digit.py

```python
numbers = [273, 103, 5, 32, 65, 9, 72, 800, 99999]

# 실행결과
# 273 은(는) 3 자릿수 입니다.
# 103 은(는) 3 자릿수 입니다.
#   5 은(는) 1 자릿수 입니다. ...


for n in numbers:
    print('{:5d} 은(는) {} 자릿수 입니다.'.format ((n), (len(str(n)))))
   
```
