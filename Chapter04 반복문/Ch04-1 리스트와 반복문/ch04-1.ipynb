> 리스트 연산자 : +, *, len()

```python
a = [1, 2, 3]
b = [4, 5, 6]
c = a + b
print(c)

d = a * 3
print(d)

print(len(d))

```

> 리스트에 요소 추가하기 : append, insert

```python
# 리스트.append(요소) : 리스트 맨 뒤에 요소가 추가됨
# 리스트.insert(위치, 요소) : 특정위치에 요소를 삽입

data1 = [1, 2, 3]
data1.append(4)
data1.append(5)
print(data1)

data = []
for i in range(30):
    data.append(i)

print(data)

data.insert(5, 'a')
print(data)
data[5]

# 리스트.extend(리스트) : 해당 리스트를 확장

li_a = ['a','b','c']
li_b = ['e','f','g']
li_c = li_a + li_b
print(li_c)

li_c.extend(['h','i'])
print(li_c)

```

> 리스트에 요소 제거하기 : 인덱스로 제거, 값으로 제거

```python
# 인덱스 사용
li = ['a','b','c','d','e','f']

# 1. del
# del li
print(li) #nameError

del li[1]
print(li)

del li[:2]
print(li)

# 2. 리스트.pop(인덱스)
print('삭제요소:', li.pop(2))
print(li)
print('삭제요소:', li.pop())
print(li)
```

```python
# 값으로 제거

# 리스트 list.remove(값)
li = [1,2,1,2]
print(li)

li.remove(2) # 리스트 내부에 해당하는 값이 여러개 있어도 가장 먼저 발견되는 하나만 제거
print(li)

li.remove(2) 
print(li)

```

```python
# 모두 제거 : clear()  del과 차이점 알아두기
li = [0,1,2,3,4,5]
print(li)

li.remove(0)
print(li)

li.clear()  
print(li)
```

> 리스트 내부에 있는지 없는지 확인하기 : in 또는 not in 연산자

```python
print('c' in 'abcdefg')

list_n = [273, 32, 103, 57]

print(273 in list_n)

print(99 not in list_n)

val = 99
if val not in list_n:
    list_n.append(val)

print(list_n)
```
