#### 시작하기전에   
프로그램은 기본적으로 '자료를 처리'하는 역할을 한다.   
* 카메라로 사진을 찍으면 사진이 **자료**이고, 이를 카메라에 저장하는 것이 **처리**이다.
* 카톡으로 메시지를 보냈다면 메시지가 **자료**이고, 이를 친구에게 전송하는 것이 **처리**이다.
* 게임에서 경험치를 얻었다면 경험치가 **자료**이고, 경험치를 증가시키는 것이 **처리**이다.

데이터를 처리한다는 것은 결국...   
" Create(생성/쓰기), Read(읽기), Update(갱신), Delete(삭제) " 한다는 것이다!!!

> 자료형(data type) 
* 문자열(string): 메일 제목, 메시지 내용 등 -> "안녕하세요", "Hello World"
* 숫자(number): 물건의 가격, 학생의 성적 등 -> 52, 273, 103.32
* 불(boolean): 로그인 상태 -> True, False   
자료형은 자료의 형태이고, 자료형에 따라 컴퓨터가 해석하는 방법이 달라진다. 

```python
print("hello")
```

> 자료형 확인하기

```python
# 파이썬 기본 자료형

print(type("나는 문자열"))
print(type(273))
print(type(3.14))
print(type(True))
print()

#  컬렉션 자료형
#  [], (), {}
print(type([1,"2", True]))
print(type({1,2,3}))
print(type({1:'a',2:'b',3:'c'}))
```

```python
# dir() : 어떤 객체가 어떤 메소드를 가지고 있는지 알고 싶을 때
dir (__builtins__)
print()
dir (10)
dir("문자열")
```

```python
#  id() : 객체를 식별하기 위해서 부여하는 일련번호
a = 10
#  10 이라는 값이 a에 담기는게 아니라 10 이라는 객체(class의 인스턴스)가 생성되고
#  참조변수 a 라는 이름으로 10 객체를 가리킨다.

id(a)

n = 4000
print("n:", id(n))

a = 4000
print("a:", id(a))
b = 4000
print("b:", id(b))

print(a == b)
print(a is b)
```

```python
#  isinstance() : 주어진 인스턴스가 특정 클래스(데이터타입)인지 검사
print(isinstance("이것은 문자열", str))
print(isinstance(2022, int))
```

```python
#  help() : 함수(function) 정보를 보고 싶을 때
# help(print)
print(print.__doc__)
```

> 문자열 만들기 : " ", '', """ """, ''' '''

```python
# 문자열('str') : 문자(char)의 배열

# 문자셋(character set)
print(ord("A")) # 인코딩
print(chr(0b1100001)) # 디코딩
print(end='\n')
for i in 'abcdef':
    print(i, end='')

```

```python
# why don't you
print('why don\'t you')
#  "Yes" they said
print('"yes" they said')
print("\"yes\" they said")
```

> 이스케이프(Escape Chracter) : \ + 문자

```python
# 특수 문자를 표현하거나 특수한 기능을 하는 문자
print('First.\nSecond Line')
# 경로명
print ('D:\MyFolder\name')
print ('D:\\MyFolder\\name')
print (r'D:\MyFolder\name')
print ('이름\t나이\t주소')
```

> 여러 줄인 문자열 생성(Multi-Line)

```python
s1 = '1. Life is short\n2. You need python'
print(s1)
#  줄바꿈 문자열
s2 = '''
1. Life is short
2. You need python
'''
print(s2)

s3 = """
**menu**
1. coffee
2. milk
3. cake
"""
print(s3)


# 줄바꿈 없는 문자열
menu = '''
1.coffee\
milk
2.cake
'''
print(menu)
```

> 문자열 연산자 : + , *

```python
# '문자열' + '문자열'
print('안녕'+'하세요')  
lan = 'python'
version = ' 3.10.1'
print(lan+version)

# '문자열' * 숫자
print('안녕하세요\n' * 3)
print('='* 50)
print('program')
print('='* 50)

```

```python
print(index)
print(array)

# 특정 단어('Life')를 추출하려면?
print(array[0:4]) # 범위표시 : 시작 <=  <끝

print(array[12:17])
print(array[:4])
print(array[19:])
print(array[:])  #처음부터 끝까지
```

```python
data = '971113-1013721'
# 생일정보, 나이정보, 성별정보 추출하시오

a = data[2:4]
b = data[4:6]
print('생일은 {}월 {}일 입니다' .format(a,b))
d = str(19) + (data[0:2])
c = 2022 - int(d)+1
print('나이는 {}세 입니다' .format(c))

if int(data[7]) == 1:
    print('성별은 남자 입니다')
else :
    print('성별은 여자 입니다')
```

```python
import datetime

idx = '01234567890123'
data = '971113-1013721'

year = data[:2]
month = data[2:4]
day = data[4:6]
print(year, month, day)

current = datetime.datetime.now()
print(current.year)
age = current.year - (1900 + int(year)) + 1






ver = int(data[7])
gender = ''

if ver % 2 == 0 :
    gender = '여성'
else :
    gender = '남성'

print('생일은 {}월 {}일 입니다.'.format(month, day))
print('나이는 {}입니다.'.format (age))
print('성별은 {}입니다.'.format (gender))
```
