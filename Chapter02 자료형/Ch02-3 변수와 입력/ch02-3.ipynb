> 복합대입연산자 : +=, -+, /=, %=, **=

```python
# 숫자 연산

a = 0
print(a+1)
a += 10 # a = a + 10
print(a)

a += 10
print(a)

a/= 5
print(a)    

# 문자열 연산(+, *)
string = '안녕하세요!'
string += '여러분'  # string = 안녕하세요! + 여러분
print(string)

ch = '~ '   
ch *= 5
print(ch)
```

> 타입 캐스팅(형변환)

```python
keyboard = input('숫자를 입력해 주세요 : ')
print('입력값', keyboard)
print(type(keyboard))

 # 'str' to 'int' : 문자열을 숫자로
keyboard = int(keyboard)
print('int(입력값):', keyboard)
print(type(keyboard))

keyboard += 10
print('입력값 + 10', keyboard)
```

> 사용자 입력값 계산하기

```python
#  숫자를 두 번 입력받아 계산하고 결과를 출력

a = int(input('첫번째 숫자를 입력하세요'))
b = int(input('두번째 숫자를 입력하세요'))

c = a+b
d = a-b
e = a*b
f = a/b

print("a + b = ", c)
print("a - b = ", d)
print("a * b = ", e)
print("a / b = ", f)
```

```python
#  숫자를 두 번 입력받아 계산하고 결과를 출력

a = int(input('첫번째 숫자를 입력하세요'))
b = int(input('두번째 숫자를 입력하세요'))

c = a+b
d = a-b
e = a*b
if a > b :
    f = a/b
else :
    f = b/a


print("a + b = ", c)
print("a - b = ", d)
print("a * b = ", e)
print("a / b = ", f)
```

> 숫자를 문자열로 바꾸기 : str()

```python
# '오늘의 날짜 : 2022. 1. 5'

import datetime

date = datetime.datetime.now()

print('오늘의 날짜 : {}. {}. {} '.format (date.year, date.month, date.day))
```

```python
import datetime
today = datetime.datetime.today().strftime('%Y-%m-%d')
today = str(today)

print(type(today))

print("오늘의 날짜는 {} 입니다" .format(today))
```
