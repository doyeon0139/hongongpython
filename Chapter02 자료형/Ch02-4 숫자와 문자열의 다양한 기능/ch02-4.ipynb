> 서식지정자(fomat specifier)로 문자열 만들기

```python
# 철수의 평균 점수는 85.3점입니다.
# 영희의 평균 점수는 75.3점입니다.
# 길동의 평균 점수는 65.3점입니다.
# 순이의 평균 점수는 95.3점입니다.
# {이름}의 평균 점수는 {점수}점입니다.


# {이름}의 평균 점수는 {점수}점입니다.

# 문자열 안에서 특정 부분을 원하는 값으로 바꿀 때 : 서식 지정자 또는 문자열 포맷팅(fomat())을 사용한다.

name =  '철수' #str  
score = 85.3    #float
print('%s의 평균 점수는 %f점입니다.' % (name, score)) #실수
print('%s의 평균 점수는 %d점입니다.' % (name, score)) #정수

name = 'IU'
print('My name is %s' % name )

```

```python
# 학급 학생들의 점수 출력하기
import time

names = ['철수','영희','길산','희경','수진','영수']
scores = [85, 98, 75, 83, 91, 88]



for i in range(len(names)) : 
    print('%s 학생의 점수는 %d점입니다.' %(names[i], scores[i]))
    time.sleep(1)

```

> format() 함수의 다양한 기능 : 숫자 관련

```python
output_a = '{:d}'.format(52)
print(output_a)


output_a = '{:s}'.format('52')
print(output_a)

output_a = '{:5d}'.format(52)
print(output_a)

output_a = '{:05d}'.format(52)
print(output_a)

print('[지출내역]')
cost1 = '{:5d} (만원)'.format(1200)
cost2 = '{:5d} (만원)'.format(200)
cost3 = '{:5d} (만원)'.format(15)
cost4 = '{:5d} (만원)'.format(3500)
cost5 = '{:5d} (만원)'.format(1200)
print(cost1)
print(cost2)
print(cost3)
print(cost4)
print(cost5)
```

```python
#  기호와 함께 출력하기
output_a = '{:+d}'.format(55)
print(output_a)

output_b = '{:+d}'.format(-55)
print(output_b)

```

```python
print('{:+5d}'.format(52))
print('{:=+5d}'.format(52))
```

```python
# 부동 소수점 출력의 형태

print('{:f}'.format(3.141592))
print('{:+15f}'.format(3.141592))

# 소수점 아래 자릿수 지정하기
print('{:15.3f}'.format(3.141592))
print('{:.5f}'.format(3.141592))

```

> 문자열 자르기 : split()

```python
# 문자열을 특정 문자로 자를 때 사용
li = '10 20 30 40 50'.split()

print(li)

fruits = 'apple pear grape pineapple orange'
li = fruits.split()
print(li)

li[0]
li[-1]

sports = 'baseball/football/golf/boxing/basketball'
li = sports.split('/')
print(li)


for s in li:
    print(s)


for i in range(len(li)):
    print(li[i], end=' ')


```

> 사용자로부터 입력 받은 모든 정수의 합을 리턴해주는 프로그램 작성 : input_split.py

```python
num = input('숫자를 모두 입력하세요')

number = num.split()

sum = 0


for i in number:
    sum += int(i)   #int(number[i]) 이게 왜 틀렸을까

print('숫자의 합은 : {} 입니다.'.format(sum))


```
