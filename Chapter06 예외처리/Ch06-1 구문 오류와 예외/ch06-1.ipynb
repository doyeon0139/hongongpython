> 기본 예외 처리(exception handling)

```python
print('프로그램 시작')

user_input = input('입력한 정수의 제곱근을 구합니다. : ')
if user_input.isdigit():
    print(f'{user_input}의 제곱근은 {int(user_input)**(1/2)} 입니다')
else:
    print('정수가 아닙니다. 프로그램을 종료합니다.')

print('프로그램 종료')
```

> try ~ except

```python
print('프로그램이 시작되었습니다.')

i = 0
PI = 3.141592

try:
    # 예외가 발생할 가능성이 있는 코드
    result = PI / i
except ZeroDivisionError as ex:
    # 예외에 해당되는 처리코드
    print('예외가 발생되었습니다.', ex)

print('프로그램이 종료되었습니다.')
```

```python
print('프로그램이 시작되었습니다.========================')

list_txt = ['52', '183', 'A', '파이썬', '103']
numbers = []
for item in list_txt:
    try:
        float(item)
        numbers.append(item)
    except Exception as ex:
        print('예외발생', str(ex))
print(numbers)

print('프로그램이 종료되었습니다.========================')
```

```python
# finally VS return


def test():
    print('test()시작')
    try:
        print('1.....')
        return
        print('2.....')
    except : print('except.....')
    else: print('else.....')
    finally: print('finally.....')
    print('text() 끝')

test()
```

```python
# finally VS break

while True:
    try:
        print('try 1.....')
        print('try 2.....')
        break
        print('try 3.....')
    except : print('except.....')
    finally:
        print('finally.....')

print('프로그램 종료')

```
